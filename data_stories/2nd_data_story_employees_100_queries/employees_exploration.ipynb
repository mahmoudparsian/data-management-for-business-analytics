## Aggregations

```
dept_agg = df.groupby('department').agg(headcount=('emp_id','count'),avg_salary=('salary','mean'),avg_age=('age','mean')).reset_index()
dept_agg.sort_values('headcount', ascending=False)
```

```
country_agg = df.groupby('country').agg(headcount=('emp_id','count'),avg_salary=('salary','mean')).reset_index()
country_agg.sort_values('headcount', ascending=False)
```

## Plots

```
# 1) Headcount by Department
ax = dept_agg.sort_values('headcount', ascending=False).plot(x='department', y='headcount', kind='bar', legend=False, title='Headcount by Department')
import matplotlib.pyplot as plt
plt.tight_layout()
```

```
# 2) Average Salary by Department
ax = dept_agg.sort_values('avg_salary', ascending=False).plot(x='department', y='avg_salary', kind='bar', legend=False, title='Average Salary by Department')
import matplotlib.pyplot as plt
plt.tight_layout()
```

```
# 3) Headcount by Country
ax = country_agg.sort_values('headcount', ascending=False).plot(x='country', y='headcount', kind='bar', legend=False, title='Headcount by Country')
import matplotlib.pyplot as plt
plt.tight_layout()
```

```
# 4) Salary distribution
ax = df['salary'].plot(kind='hist', bins=10, title='Salary Distribution')
import matplotlib.pyplot as plt
plt.tight_layout()
```

```
# 5) Age vs Salary
ax = df.plot(x='age', y='salary', kind='scatter', title='Age vs Salary')
import matplotlib.pyplot as plt
plt.tight_layout()
```
